```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
project_root = Path.cwd().resolve().parents[0]
project_root
```

```python
# loading processed yield curves
file_path = project_root / 'data' / 'processed' / 'cleaned_yield_curve_fred.csv'

df = pd.read_csv(
    file_path,
    index_col = 'date',
    parse_dates = True
)

plt.figure(figsize=(14, 7))

for col in df.columns:
    plt.plot(
        df.index, 
        df[col], 
        label = col
    )

plt.legend(loc='upper right', title = 'Series')
plt.title('Yield Curve Series', fontweight= 'bold')
plt.show()
```
